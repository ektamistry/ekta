Calculate HOG

```python
from skimage.feature import hog
```

For Object 1

Pixels per cell = (8x8)

```python
fd, hog_image1 = hog(obj1, orientations=9, pixels_per_cell=(8, 8),
                    cells_per_block=(2, 2), visualize=True, multichannel=True)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8,10), sharex=True, sharey=True)

ax1.axis('off')
ax1.imshow(obj1, cmap=plt.cm.gray)
ax1.set_title('Object 1')
ax2.axis('off')
ax2.imshow(hog_image1, cmap=plt.cm.gray)
ax2.set_title('HOG of Object 1')
plt.show()
```

Pixels per Cell = (16x16)

```python
fd, hog_image12 = hog(obj1, orientations=9, pixels_per_cell=(16, 16),
                    cells_per_block=(2, 2), visualize=True, multichannel=True)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8,10), sharex=True, sharey=True)

ax1.axis('off')
ax1.imshow(obj1, cmap=plt.cm.gray)
ax1.set_title('Object 1')
ax2.axis('off')
ax2.imshow(hog_image12, cmap=plt.cm.gray)
ax2.set_title('HOG of Object 1')
plt.show()
```

Image Similarity between Object 1 and Object 2

```python
from sklearn.metrics.pairwise import euclidean_distances
import numpy as np
```

```python
# Convert images to numpy arrays
image1_array = np.array(hog_image1)
image2_array = np.array(hog_image2)
```

```python
# Reshape image arrays to 2D arrays (flatten the images)
flat_image1 = image1_array.reshape(-1, image1_array.shape[-1])
flat_image2 = image2_array.reshape(-1, image2_array.shape[-1])
```

```python
# Compute the pairwise Euclidean distances between pixel values
distances = euclidean_distances(flat_image1, flat_image2)
```

```python
distances
```

```python
similarity_scores = 1 / (1 + distances)
print(similarity_scores)
```

```python
# Calculate the average similarity score
average_similarity_score = np.mean(similarity_scores)
average_similarity_score*100
```

**Conclusion** : For the given image hog features and hog image are determined for the following set of parameters: pixel_per_cells = (8,8), cells_per_block=(2,2), orientations=9. For the same image number of pixels per cells is increased to (16,16) and reduced to (4,4). It is observed that increase in cell size results in broader details on hog of image and for smaller cell size finer details on the hog of the image. Each cell of hog of image carries a line. Length and intensity of line represents the magnitude of gradient of the pixels in that cell. And orientation of the lines represents the direction of the edge within that cell. If a particular cell is dark then cell does not have edge. The euclidean distance between two similar objects is less than two dissimilar objects, which shows that first two objects are similar.
